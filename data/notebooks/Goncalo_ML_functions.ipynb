```python
import pandas as pd
import numpy as np

# load cleaned data
df = pd.read_csv('../clean/goncalo_clean.csv')

# safety copy
df = df.copy()

df.head()
```

The goal of this step is to create new features that better capture patterns in the data and improve the model’s ability to predict movie popularity.


```python
df['popularity_log'] = np.log1p(df['popularity'])
```

```python
df['runtime_category'] = pd.cut(
    df['runtime'],
    bins=[0, 90, 120, 150, 240],
    labels=['Short', 'Standard', 'Long', 'Very Long']
)
```

```python
df['runtime_squared'] = df['runtime'] ** 2
```

```python
df['runtime_x_rating'] = df['runtime'] * df['vote_average']
```

```python
df['rating_category'] = pd.cut(
    df['vote_average'],
    bins=[0, 5, 6.5, 8, 10],
    labels=['Low', 'Average', 'Good', 'Excellent']
)
```

Several features were engineered to improve the model’s predictive power. A log transformation was applied to the popularity variable to reduce skewness and stabilize variance. Polynomial and interaction features were created to capture non-linear relationships between runtime and popularity, as well as the combined effect of runtime and movie ratings. Additionally, runtime and rating were grouped into categories to help the model capture broader patterns that may not be reflected in raw numerical values.
